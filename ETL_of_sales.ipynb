```python
import pandas as pd
```

```python
df = pd.read_csv("sales_data.csv")
```

```python
df.dropna(inplace = True)
df.drop_duplicates(inplace = True)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Revenue'] = df['Unit Cost'] * df['Unit Price']
```

```python
df.head()
```

```python
df.to_csv("Clean Sales Data.csv", index= False)
```
